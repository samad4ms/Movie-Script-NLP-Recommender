```python
# show topics and pertaining words

def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
```

```python
# clean script with text processing and split. No parts of speech yet.


def clean_split(script):

    # covert script to string
    script_ = str(script)
    
    script_ = script_.replace('-'," ")

    # remove punctuation
    script_ = re.sub('[%s]' % re.escape(string.punctuation),"", script_)

    # lower case everything
    script_ = script_.lower()

    # remove numbers here
    script_ = re.sub('\w*\d\w*', ' ', script_)
    
    # replace characters I want to split on
    # n = 1000
    # script_ = [script_[i:i+n] for i in range(0, len(script_), n)]
    
    chunks, chunk_size = len(script_), len(script_)//99
    script_ = [script_[i:i+chunk_size] for i in range(0, chunks, chunk_size)]
    
    return script_
```

```python
# lemmatize, tokenize and remove stops

def tokenize(split_script):
    
    text_ = str(split_script)
    
    # lemmatization 
    lemmatizer = WordNetLemmatizer()
    text_ = lemmatizer.lemmatize(text_)
    
    # tag each word with part of speech
    tokens = pos_tag(word_tokenize(text_))
    
    # isolote only adj and verbs, nouns    ,'NN','NNS'
    pos = ['JJ','JJR','JJS','RB','RBR','RBS','VB','VBG','VBD','VBN','VBP','VBZ','RP']
    verbs_adj = [i[0] for i in tokens if i[1] in pos]
    
    # stop words
    stops_apost = stopwords.words('english') + names_all + movie_term_stops + others + names_plural
    stops = [i.replace("'", '') for i in stops_apost]
    tokens_stop = [i for i in verbs_adj if i not in stops]
    
    return tokens_stop
```

```python
# final recommender

def recommend_movie(fav_movie):
    sort_input = distances_named[fav_movie].sort_values(ascending=False)
    mask = sort_input.index.isin([fav_movie])
    all_results = sort_input[~mask]
    
    return all_results[:4] # return 4 most similar movies

# run recommeder function
recommend_movie('Taxi-Driver')
```

# LDA

```python
stops_apost = stopwords.words('english') + names_all + movie_term_stops + others + names_plural
stops = [i.replace("'", '') for i in stops_apost]
```

```python
# Create a CountVectorizer for parsing/counting words
count_vectorizer = CountVectorizer(stop_words = stops, token_pattern="\\b[a-z][a-z]+\\b", max_df=0.6, min_df=3)

# Create the same term-document matrix but TRANSPOSE. (word as rows)
doc_word_lda = count_vectorizer.fit_transform(all_movies_split).transpose()

doc_word_lda.shape
```

```python
# Convert sparse matrix of counts to a gensim corpus
corpus = matutils.Sparse2Corpus(doc_word_lda)
```

```python
## Map matrix rows to words (tokens)

# We need to save a mapping (dict) of 
# row id to word (token) for later use by gensim:

id2word = dict((v,k) for k,v in count_vectorizer.vocabulary_.items())

len(id2word)
```

```python
# Create lda model (equivalent to "fit" in sklearn)
lda = models.LdaModel(corpus=corpus, num_topics=5, id2word=id2word, passes=10)
```

```python
#see what happened. Top 5 most important words in each topics. 

lda.print_topics()
```

If we want to map our documents to the topic space we need to actually use the LdaModel transformer that we created above, like so:

```python
# Transform the docs from the word space to the topic space (like "transform" in sklearn)

lda_corpus = lda[corpus]
lda_corpus
```

```python
# Store the documents' topic vectors in a list so we can take a peak
lda_docs = [doc for doc in lda_corpus]
```
